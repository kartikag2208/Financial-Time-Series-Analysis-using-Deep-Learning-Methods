```python
### Dheeraj M

#### Reference of implementation taken from : https://github.com/drimpossible/GDumb



from train import *

class MemoryBuffer:
    def __init__(self, buffer_size):
        self.buffer_size = buffer_size
        self.data = []
        self.labels = []

    def add_samples(self, samples, labels):
        
        self.data.extend(samples)
        self.labels.extend(labels)
      
        if len(self.data) > self.buffer_size:
            indices = np.random.choice(len(self.data), self.buffer_size, replace=False)
            self.data = [self.data[i] for i in indices]
            self.labels = [self.labels[i] for i in indices]

    def get_data(self):
        return torch.cat(self.data, dim=0), torch.cat(self.labels, dim=0)
    
def train_gdumb(model_class, buffer, device, epochs=20, batch_size=32, learning_rate=0.001):
    # Re-initialize the model
    model = model_class(input_size=1).to(device)
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)

    # Prepare buffer data
    X, y = buffer.get_data()
    train_loader = torch.utils.data.DataLoader(
        torch.utils.data.TensorDataset(X.to(device), y.to(device)), batch_size=batch_size, shuffle=True)

    model.train()
    for epoch in range(epochs):
        total_loss = 0
        for x_batch, y_batch in train_loader:
            optimizer.zero_grad()
            preds = model(x_batch)
            loss = criterion(preds, y_batch)
            loss.backward()
            optimizer.step()
            total_loss += loss.item()
        avg_loss = total_loss / len(train_loader)
        print(f"Epoch [{epoch+1}/{epochs}], Loss: {avg_loss:.4f}")

    return model
```
