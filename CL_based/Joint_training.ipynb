```python
import time
import torch
from torch.utils.tensorboard import SummaryWriter

input_size = 1
hidden_size = 64
num_layers = 2
output_size = 1

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

ticker_label = "TATAMOTORS.NS"
data_periods = []
START = "2000-08-01"
END = "2010-12-31"

data_periods.append(("2010-12-31", "2013-12-31"))
data_periods.append(("2013-12-31", "2016-12-31"))
data_periods.append(("2016-12-31", "2019-12-31"))
data_periods.append(("2019-12-31", "2022-12-31"))

# Record simple metrics
metrics = []

dataset_history, scaler, test_dates = dataset_extraction(
    stock=ticker_label, start_time=START, stop_time=END, sequence_length=60
)
model = LSTMModel(input_size, hidden_size, num_layers, output_size).to(device)

# Track time for initial training
start_time = time.time()
Model = training(dataset=dataset_history, Model=model, scaler=scaler, test_dates=test_dates)
end_time = time.time()
metrics.append({
    "data_period": f"{START} to {END}",
    "training_time": end_time - start_time,
    "peak_memory": torch.cuda.max_memory_allocated() / 1e6 if torch.cuda.is_available() else "N/A"
})
torch.cuda.reset_peak_memory_stats()

data = dataset_history

for i, (start, end) in enumerate(data_periods):
    model = LSTMModel(input_size, hidden_size, num_layers, output_size).to(device)
    
    data_new, scaler, test_dates = dataset_extraction(
        stock=ticker_label, start_time=start, stop_time=end, sequence_length=60, train_ratio=0.7, val_ratio=0.1
    )
    if i != 0:
        stop = data_periods[i - 1][1]
        data, _, _ = dataset_extraction(
            stock=ticker_label, start_time=START, stop_time=stop, sequence_length=60
        )
    data = time_dataset_concatenate(data, data_new)

    train_count, val_count, test_count, total_count, start_test_date, end_test_date = dataset_summary(data, test_dates)

    # Time and memory tracking for current period
    start_time = time.time()
    Model = training(dataset=data, Model=model, scaler=scaler, test_dates=test_dates)
    end_time = time.time()

    metrics.append({
        "data_period": f"{start} to {end}",
        "training_time": np.round(end_time - start_time,2),
        "peak_memory": torch.cuda.max_memory_allocated() / 1e6 if torch.cuda.is_available() else "N/A"
    })
    torch.cuda.reset_peak_memory_stats()
    del model
    torch.cuda.empty_cache()

# Print summary metrics
print("Summary Metrics:")
for metric in metrics:
    print(metric)
```
